```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import tensorflow as tf
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report


#  Import and read the charity_data.csv.
import pandas as pd
spotify_df = pd.read_csv("/content/drive/MyDrive/Project_4/resources/Popular_Spotify_Songs.csv", encoding='ISO-8859-1')
spotify_df
```

```python
spotify_df.info()
```

```python
#remove unnessasary data
cleaned_spotify_df = spotify_df.drop(["in_apple_playlists","in_apple_charts","in_deezer_playlists","in_deezer_charts","in_shazam_charts"], axis=1)
```

```python
# Replace or remove non-numeric values in the 'streams' column
# For instance, if the non-numeric value is consistent, you can replace it:
cleaned_spotify_df['streams'] = cleaned_spotify_df['streams'].replace('BPM110KeyAModeMajorDanceability53Valence75Energy69Acousticness7Instrumentalness0Liveness17Speechiness3', '0')

# Convert the 'streams' column to integers
cleaned_spotify_df['streams'] = cleaned_spotify_df['streams'].astype(int)# Check for unique values in the 'streams' column
cleaned_spotify_df.head()
```

```python
cleaned_spotify_df.info()
```

```python
# encoding key and mode into numerical values
cleaned_spotify_df['key']=LabelEncoder().fit_transform(cleaned_spotify_df['key'])
cleaned_spotify_df['mode']=LabelEncoder().fit_transform(cleaned_spotify_df['mode'])
cleaned_spotify_df.head()
```

# **Training and Test sets for predicting streams**

```python
#splitting the data into training and testing sets
training_data, testing_data = train_test_split(cleaned_spotify_df, test_size=0.1, random_state=30)
print(f"No. of training examples: {training_data.shape[0]}")
print(f"No. of testing examples: {testing_data.shape[0]}")
testing_data.head()
```

```python
#dropping streams column because that is what we are trying to predict
testing_data_streams = testing_data['streams'] #Save the 'streams' column before dropping it
testing_data = testing_data.drop('streams', axis=1) #Drop the 'streams' column

#setting up x train and y train
x_train = training_data.drop(['streams', 'track_name', 'artist(s)_name'], axis=1) #Drop both 'streams','track_name' and 'artist(s)_name'
y_train=training_data['streams']
x_train.shape, testing_data.shape, y_train.shape
x_train.head()
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
# Fitting Standard Scaller
X_scaler = scaler.fit(x_train)
# Scaling data
X_train_scaled = X_scaler.transform(x_train)
```

# **Random Forrest for streams**




```python
#Modelling and training
rf_model = RandomForestRegressor(n_estimators=50)
rf_model.fit(X_train_scaled,y_train)
print(rf_model.score(X_train_scaled,y_train))
```

```python
#making predictions on test data
#Drop the 'track_name' column from testing_data
testing_data = testing_data.drop('track_name', axis=1)
#Drop the 'artist(s)_name' column from testing_data
testing_data = testing_data.drop('artist(s)_name', axis=1)
rf_predict = rf_model.predict(testing_data)
```

```python
# 'testing_data_streams' contains the true labels and you want to
# evaluate the regression model, you need to use appropriate regression metrics.

from sklearn.metrics import mean_squared_error, r2_score

# Calculate Mean Squared Error
mse = mean_squared_error(testing_data_streams, rf_predict)
print("Mean Squared Error:", mse)

# Calculate R-squared
r2 = r2_score(testing_data_streams, rf_predict)
print("R-squared:", r2)
```

```python
#outputting the predictions
rf_output = pd.DataFrame({'Predictions': rf_predict, 'Actual' : testing_data_streams})
rf_output.head()
```

# **Linear Regression for streams**

```python

#Modelling and Training
linear_model = LinearRegression()
linear_model.fit(X_train_scaled,y_train)
print(linear_model.score(X_train_scaled,y_train))

#testing the model
linear_predict = linear_model.predict(testing_data)
l_output = pd.DataFrame({'Predictions': linear_predict, 'Actual' : testing_data_streams})
l_output.head()
```
